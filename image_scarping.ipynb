```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
import pandas as pd
```

```python
socities = pd.read_csv('recommendation_data.csv')['PropertyName'].tolist()
```

```python
socities
```

```python
firefox_options = Options()
firefox_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")
cookies = [
    {'name': 'cookie_name_1', 'value': 'cookie_value_1'},
    {'name': 'cookie_name_2', 'value': 'cookie_value_2'},
]
driver = webdriver.Firefox(options=firefox_options)
for cookie in cookies:
    driver.add_cookie(cookie)
```

```python
def get_google_photos_url(query):
    try:
        driver.get("https://www.google.com/imghp")
        search_box = driver.find_element(By.NAME, "q")
        search_box.send_keys(query)
        search_box.send_keys(Keys.RETURN)

        # Use WebDriverWait to wait for the first image to be present on the page
        timeout = 10  # seconds
        WebDriverWait(driver, timeout).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "img[jsname='Q4LuWd']"))
        )

        # Find the first image element
        first_image = driver.find_element(By.CSS_SELECTOR, "img[jsname='Q4LuWd']")
        image_url = first_image.get_attribute("src")
        return image_url

    except TimeoutException:
        print(f"Timed out waiting for the first image to appear on the page.")
```

```python
society_images = {}
for society_name in socities:
    print(society_name) # for debugging
    query = society_name + ' society, Gurgaon best image'
    society_images[society_name] = get_google_photos_url(query)
```
