```python
# 1 bhk, 2 bhk, 3 bhk, 4 bhk, 5 bhk, land, villa, independent floor, studio
def extract_subname_details(subname):
    subname = subname.split('in')[0].replace(',', ' ').lower()
    result = [0] * 9
    
    if '1' in subname:
        result[0] = 1
    if '2' in subname:
        result[1] = 1
    if '3' in subname:
        result[2] = 1
    if '4' in subname:
        result[3] = 1
    if '5' in subname:
        result[4] = 1
    if 'land' in subname:
        result[5] = 1
    if 'villa' in subname:
        result[6] = 1
    if 'independent floor' in subname:
        result[7] = 1
    if 'studio' in subname:
        result[8] = 1
    return result
```

```python
data[['1_bhk', '2_bhk', '3_bhk', '4_bhk', '5_bhk', 'land', 'villa', 'independent_floor', 'studio']] = np.array(data['PropertySubName'].apply(extract_subname_details).tolist())
```

```python
import re
def get_distance(distance):

    distance = distance.lower().replace('mtr', 'meter').replace('km', ' km')
    if 'meter' in distance:
        pattern = r'\b\d*\.\d+|\b\d+\b'
        matches = re.findall(pattern, distance)
        if len(matches) == 0:
            print(distance, '1')
        return float(matches[0])
    elif 'km' in distance:
        pattern = r'\b\d*\.\d+|\b\d+\b'
        matches = re.findall(pattern, distance)
        if len(matches) == 0:
            print(distance, '2')
        return float(matches[0]) * 1000
    elif 'within reach' in distance or 'close proximity' in distance or 'in proximity' in distance or 'closeby' in distance or 'in close proximity' in distance or 'located nearby' in distance or 'nearby' in distance:
        return 500.0
    elif 'within' in distance:
        distance = distance.replace('m', ' m')
        pattern = r'\b\d*\.\d+|\b\d+\b'
        matches = re.findall(pattern, distance)
        if len(matches) == 0:
            print(distance, '3')
        return float(matches[0])
    elif 'min' in distance:
        average_speed = 500 # 500 m/min or 30 kmph
        distance = distance.replace('m', ' m')
        pattern = r'\b\d*\.\d+|\b\d+\b'
        matches = re.findall(pattern, distance)
        if len(matches) == 0:
            print(distance, '4')
        return float(matches[0]) * average_speed
    elif 'm' in distance:
        distance = distance.replace('m', ' m')
        pattern = r'\b\d*\.\d+|\b\d+\b'
        matches = re.findall(pattern, distance)
        if len(matches) == 0:
            print(distance, '5')
        return float(matches[0])
    else:
        return float(distance.strip()) * 1000 # since only 1.4 was left
```

```python
def clean_location_advantage(location_advantages):
    result = [50000.0] * len(unique_locations)
    
    for location, distance in location_advantages.items():
        original_location = get_original_name(location)
        cleaned_distance = round(get_distance(distance), 2)

        for i in range(len(unique_locations)):
            loc = unique_locations[i]
            if loc == original_location:
                result[i] = cleaned_distance
    return result
```

```python
def get_similarity(property_name, data, top_n=5):
    data = data.set_index('PropertyName')
    scaler = StandardScaler()
    data = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
    data.index = index_
    features = data.loc[property_name].values.reshape(1, -1)

    similarities = cosine_similarity(features, data.values)
    return similarities
```

```python
sector_columns = ['PropertyName'] + data.columns[1: 62+1].tolist() 
building_details_columns = ['PropertyName'] + data.columns[63: 71+1].tolist()
nearby_locations_columns = ['PropertyName'] + data.columns[72: 784+1].tolist()
facilities_columns = ['PropertyName'] + data.columns[786:].tolist()
```

```python
def get_top_k_similar_societies(property_name, top_n=5):
    sector_similarity = get_similarity(property_name, data[sector_columns], top_n)
    building_details_similarity = get_similarity(property_name, data[building_details_columns], top_n)
    nearby_locations_similarity = get_similarity(property_name, data[nearby_locations_columns], top_n)
    facilities_similarity = get_similarity(property_name, data[facilities_columns], top_n)
    
    sector_weightage = 8
    building_details_weightage = 6
    nearby_locations_weightage = 5
    facilities_weightage = 3
    
    net_similarity = sector_weightage * sector_similarity + \
                    building_details_weightage * building_details_similarity + \
                    nearby_locations_weightage * nearby_locations_similarity + \
                    facilities_weightage * facilities_similarity
    
    top_indices = net_similarity.argsort()[0][::-1][1:top_n+1]

    top_similar_flats = data.loc[top_indices, 'PropertyName'].tolist()
    return top_similar_flats
```
